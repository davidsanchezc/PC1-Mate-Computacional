```python
try:
    objEntrada = open("notas.txt","r")
    objSalida = open("notas2.txt","w")
    linea=objEntrada.readline()
    while linea!="":
        alumno=linea.split(" ")
        
        print(alumno[0]+" | "+alumno[1]+" | "+alumno[2])
        objSalida.write(alumno[0]+" "+alumno[1]+" "+alumno[2])
        linea=objEntrada.readline()
    objEntrada.close()
    objSalida.close()
except FileNotFoundError:
    print('Â¡No se puede abrir el archivo especificado!')
```
